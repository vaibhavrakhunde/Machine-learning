```python
import spacy
from spacy import displacy
import pandas as pd
```

```python
nlp  = spacy.load('en_core_web_sm')

nlp.pipe_names
```

```python
doc = nlp('Mr.Elon is going to acquire Twitter Inc for $45 billion in 2023')
```

```python
doc.ents
```

```python
for i in doc.ents:
    print(i, "|", i.label_)
```

```python
displacy.render(doc, style = 'ent')
```
